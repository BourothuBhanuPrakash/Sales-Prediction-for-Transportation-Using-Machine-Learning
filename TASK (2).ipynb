# 2)Correlation Analysis

```python
colormap = plt.cm.viridis
plt.figure(figsize = (12,12))
plt.title('correltion', y=1.05, size=15)
sns.heatmap(data.corr(),linewidth=0.1,vmax=1.0,square=True,cmap=colormap,linecolor='white',annot=True)
```

### Based on the correlation heatmap provided, we can identify which variables have the strongest association with the rate variables (`target_rate`, `avg_rate`, and `max_rate`) based on their correlation values.

1. Strong Correlation with Rate Variables:
   - target_rate, avg_rate, and max_rate are highly correlated with each other, with correlation values close to 1.0. This is expected as they likely represent similar or related metrics.
   
2. Moderate Correlation with Origin and Destination Zips:
   - origin_zip and destination_zip show moderate correlations with the rate variables:
     - `origin_zip` has a correlation of 0.28 with target_rate, avg_rate, and max_rate.
     - `destination_zip` has a slightly lower correlation (around 0.14-0.15) with the rate variables.
   
3. Other Variables:
   - miles also shows a moderate positive correlation (around 0.7) with the rate variables, indicating that the distance between origin and destination might influence the rate.


- origin_zip and destination_zip are moderately associated with the rate variables, but other factors, such as `miles`, also play a significant role.
- Among these, origin_zip has a slightly stronger association with the rate variables than destination_zip.

Based on this heatmap, origin_zip and miles would be important factors to consider when analyzing rates based on location.
